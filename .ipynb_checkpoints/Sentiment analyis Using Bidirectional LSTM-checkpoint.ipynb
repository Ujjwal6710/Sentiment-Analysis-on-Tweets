```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
print(tf.__version__)
# Download all ntlk resources
import nltk
# nltk.download('all')

from keras_preprocessing.text import Tokenizer
from keras.layers import Dense, Dropout, Embedding, LSTM, Bidirectional,InputLayer
from sklearn.model_selection import train_test_split
from nltk.stem import WordNetLemmatizer
import warnings
import keras
import re
from nltk.corpus import stopwords

## Wordcloud
from wordcloud import WordCloud

warnings.filterwarnings("ignore") 
```

```python
df = pd.read_csv('Dataset_Sentiment_Analysis.csv', encoding = "ISO-8859-1",names = ['polarity','id','date','query','user','text'])
```

```python
df.head()
```

```python
df = df.drop(['id','date','query','user'],axis=1)
```

```python
df = df.sample(frac=1)
df = df[:100000]
```

```python
df['polarity'].value_counts()
```

```python
df['polarity'] = df['polarity'].replace(4,1)
```

```python
df['text'] = df['text'].str.lower()
```

```python
df['text']=df['text'].apply(lambda x:re.sub(r"((http://)[^ ]*|(https://)[^ ]*|( www\.)[^ ]*)", '',x))
df['text']=df['text'].apply(lambda x: re.sub(r'@[^\s]+', '',x))
df['text']=df['text'].apply(lambda x: re.sub(r'#', '',x))
df['text']=df['text'].apply(lambda x: re.sub(r'--+', '',x))
df['text']=df['text'].apply(lambda x: re.sub(r'/', '',x))
df['text']=df['text'].apply(lambda x: re.sub(r'_', '',x))
## Remove the stopswords
df['text']=df['text'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Remove url 
df['text']=df['text'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))

df['text']=df['text'].apply(lambda x:re.sub('[^a-z A-z 0-9-]+', '',x))
## Remove any additional spaces
df['text']=df['text'].apply(lambda x: " ".join(x.split()))
```

```python
df['text']
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['text']=df['text'].apply(lambda x:lemmatize_words(x))
```

```python
df['text']
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
X = np.array(df['text'])
```

```python
X.shape
```

```python
y= np.array(df['polarity'])
```

```python
y.shape
```

```python
# Count total no. of distinct tokens

tokenizer = Tokenizer(filters='@')
tokenizer.fit_on_texts(X)

print('No. of distinct tokens = '+str(len(tokenizer.word_index)))
```

```python
# Define Vocabulary size (no. of most frequent tokens) to consider

max_vocab=50000
```

```python
# Reload Twitter dataset with new Vocabulary

tokenizer = Tokenizer(num_words=max_vocab,filters='@')
tokenizer.fit_on_texts(X)
```

```python
# Vectorize input text using Vocabulary

X_vectorized=tokenizer.texts_to_sequences(X)
```

```python
# Count average length of tweets

sum=0
for sentence in X_vectorized:
  sum+=len(sentence)

print('Average length of tweets = '+str(sum/len(X_vectorized)))
```

```python
# Define Maximum input length of the Model

max_length=15
```

```python
# Pad or Trim data to defined input length

X_pad = keras.preprocessing.sequence.pad_sequences(X_vectorized,max_length,padding='post',
                                                   truncating='post')

print(X_pad.shape)
```

```python

model = keras.models.Sequential()

model.add(InputLayer(input_shape=(15,)))
model.add(Embedding(max_vocab, 100,input_length=max_length,mask_zero=True,
                      trainable=False))
model.add(Bidirectional(LSTM(64)))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer,loss='binary_crossentropy',
                metrics=['accuracy'])
```

```python
model.summary()
```

```python
np.random.seed(123)

X_train, X_test, Y_train, Y_test = train_test_split (X_pad,y.reshape(y.shape[0],1),test_size=0.05)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Train the model

history = model.fit(X_train,Y_train,batch_size=1024,epochs=15,validation_split=0.05)
```

```python
# Evaluate model on Test data

model.evaluate(X_test,Y_test)
```

```python
y_pred=model.predict(X_test)
```

```python
y_pred = np.where(y_pred > 0.6, 1,0)
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score, classification_report
```

```python
print(classification_report(Y_test,y_pred))
```

```python
print(confusion_matrix(Y_test,y_pred))
```
